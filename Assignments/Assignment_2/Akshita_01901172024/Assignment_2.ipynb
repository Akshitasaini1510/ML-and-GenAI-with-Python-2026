1. Find sum of first 10 natural numbers


```python
sum = 0
for i in range(1,11):
    sum+=i
print("The sum of first 10 natural numbers is: ",sum)
```

2. Find factorial of a number.

```python
num = int(input("Enter the number: "))
fac = 1;
for i in range(2,num+1):
    fac*=i
print("The factorial of number is: ",fac)
```

3. Print Fibonacci Series

```python
num = int(input("Enter the number of elements you want to print in the Fibonacci Series: "))
arr = [0,1]
for i in range(1,num-1):
    arr.append(arr[i]+arr[i-1])
print(arr)
```

4. Find largest among 3 numbers

```python
n1 = int(input("Enter first number: "))
n2 = int(input("Enter second number: "))
n3 = int(input("Enter three numbers: "))
if(n1 > n2 and n1> n3):
    print(n1)
elif(n2 > n3):
    print(n2)
else:
    print(n3)
```
